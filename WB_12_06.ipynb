### Minimising number of empty 'Size' cells

```python
import re

# Function to extract sizes

# Define regex patterns for various size formats
patterns = [
    r'\b\d+ л\b',       # liters
    r'\b\d+мл\b',     # milliliters
    r'\b\d+ г\b',       # grams
    r'\b\d+ гр\b',     # grams (alternative)
    r'\b\d+ кг\b',     # kilograms
    r'\b\d+ штуки\b',      # pieces
    r'\b\d+шт\b'      # pieces
]

def extract_size(name):
    matches = []
    # Search for each pattern in the name
    for pattern in patterns:
        match = re.search(pattern, name)
        if match:
            matches.append(match.group())

    # Join all matches if found, else return 'n/a'
    if matches:
        return ', '.join(matches)
    else:
        return 'n/a'

# Apply the function to the empty cells in the 'Size' column
df.loc[df['Size'].isna(), 'Size'] = df.loc[df['Size'].isna(), 'Name'].apply(extract_size)
```
