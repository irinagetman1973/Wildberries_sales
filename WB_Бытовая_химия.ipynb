```python
# Function to categorize rows based on keywords
def categorize(row):
    combined_text = str(row['Name']) + ' ' + str(row['Subject']) + ' ' + str(row['Category'])
    matched_categories = []
    for category, keywords in categories.items():
        if any(keyword.lower() in combined_text.lower() for keyword in keywords):
            matched_categories.append(category)
    return ', '.join(matched_categories) if matched_categories else ""

# Apply the categorization
df['Гетьман'] = df.apply(categorize, axis=1)
```

```python
import re

# Function to extract sizes
def extract_size(name):
    # Define regex patterns for various size formats
    patterns = [
        r'\b\d+л\b',       # liters
        r'\b\d+ мл\b',     # milliliters
        r'\b\d+г\b',       # grams
        r'\b\d+ гр\b',     # grams (alternative)
        r'\b\d+ кг\b',     # kilograms
        r'\b\d+ шт\b'      # pieces
    ]

    # Search for each pattern in the name
    for pattern in patterns:
        match = re.search(pattern, name)
        if match:
            return match.group()
    return None

# Apply the function to the Name column
df['Size'] = df['Name'].apply(extract_size)
```

```python
# Reorder columns to place 'Size' after 'Name' and 'Subcategory' after 'Size'
cols = df.columns.tolist()
name_index = cols.index('Name')  # Find the index of the 'Name' column

# Construct new column order
new_cols_order = cols[:name_index + 1] + ['Size', 'Subcategory'] + cols[name_index + 1:]

# Reorder DataFrame columns
df = df[new_cols_order]

```

```python
# Remove duplicate columns
df = df.loc[:, ~df.columns.duplicated()]
```

```python
# Define the subcategories for each major category
subcategories = {
    'ПММ': {
        'таблетки': ['таблетки', 'табл'],
        'порошки': ['порошок', 'порошки'],
        'гель': ['гель', 'гели'],
        'соль': ['соль', 'соли'],
        'ополаскиватель': ['ополаскиватель', 'ополаскиватели']
    },
    'Чистящие кремы/спреи/жидкости': {
        'кремы': ['кремы', 'крем'],
        'спреи': ['спреи', 'спрей'],
        'жидкости': ['жидкости', 'жидкость']
    },
    'Чистящие порошки': {
        'порошки': ['порошки', 'порошок', 'средство']
    }
}

# Function to classify subcategories
def classify_subcategory(row):
    major_category = row['Гетьман']
    combined_text = str(row['Name']).lower() + ' ' + str(row['Subject']).lower()

    if major_category in subcategories:
        for subcategory, keywords in subcategories[major_category].items():
            if any(keyword in combined_text for keyword in keywords):
                return subcategory
    return 'other'  # Default subcategory if no match is found

# Apply the function to classify subcategories
df['Subcategory'] = df.apply(classify_subcategory, axis=1)

# Display the DataFrame to check the results
df[['SKU', 'Name', 'Size', 'Subcategory', 'Гетьман', 'Subject', 'Category']]
```
